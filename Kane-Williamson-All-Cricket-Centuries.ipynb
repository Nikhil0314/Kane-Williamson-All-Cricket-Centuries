```python
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')

df = df.dropna(subset=['Score'])
```

```python
df
```

```python
#Distribution of Kane Williamson Century Scores
```

```python
plt.figure(figsize=(10, 6))
df['Score'].plot(kind='hist', bins=10, edgecolor='black')
plt.xlabel('Century Score')
plt.ylabel('Frequency')
plt.title('Distribution of Kane Williamson\'s Century Scores')
plt.show()
```

```python
#Kane Williamson\'s Centuries Against Different Opponents
```

```python
opponent_counts = df['Against'].value_counts()
plt.figure(figsize=(8, 8))
opponent_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Kane Williamson\'s Centuries Against Different Opponents')
plt.ylabel('')
plt.show()
```

```python
#Kane Williamson\'s Centuries Against Different Opponents
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df.sort_values('Date', inplace=True) 
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Score'], marker='o')
plt.xlabel('Date')
plt.ylabel('Century Score')
plt.title('Kane Williamson\'s Centuries Against Different Opponents')
plt.xticks(rotation=45)
plt.show()
```

```python
#Relationship between Batting Position and Century Scores
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Position'], df['Score'])
plt.xlabel('Batting Position')
plt.ylabel('Century Score')
plt.title('Relationship between Batting Position and Century Scores')
plt.show()
```

```python
#Distribution of Kane Williamson's Century Scores by Batting Position
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='Score', by='Position')
plt.xlabel('Batting Position')
plt.ylabel('Century Score')
plt.title("Distribution of Kane Williamson's Century Scores by Batting Position")
plt.show()
```

```python
#Number of Kane Williamson's Centuries by Opponent and Result
```

```python
opponent_result = df.groupby(['Against', 'Result']).size().unstack()
opponent_result.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Opponent')
plt.ylabel('Number of Centuries')
plt.title("Number of Kane Williamson's Centuries by Opponent and Result")
plt.legend(title='Result', loc='upper right')
plt.show()
```

```python
#Cumulative Century Scores of Kane Williamson over Time
```

```python
df['Cumulative Score'] = df['Score'].cumsum()
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Cumulative Score'], marker='o')
plt.xlabel('Date')
plt.ylabel('Cumulative Century Score')
plt.title("Cumulative Century Scores of Kane Williamson over Time")
plt.xticks(rotation=45)
plt.show()
```
